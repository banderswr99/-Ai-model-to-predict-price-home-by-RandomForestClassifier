```python
import pandas as pd
#فصل البيانات
from sklearn.model_selection import train_test_split as sp
#نموذج الذكا الاصطناعي اللي بنستخدمه
from sklearn.ensemble import RandomForestClassifier as ai



#reading file huod data
data = pd.read_csv('housing_data_large.csv')

x=data[['space','rooms','bathrooms','age','district_code',]]
y=data['price']
# x المدخلات y المخرجات  .   ،، التست سايز نسبه العينه التدريب من البيانات 
xTrain,xTest,yTrain,yTest = sp(x,y,test_size=0.3,random_state=42)
#  n_estimators=42,random_state=42
model=ai(n_estimators=100,random_state=42)
model.fit(xTrain,yTrain)
pp=model.predict(xTest)






sample=[[350,5,4,3,104]]
hh=model.predict(sample)
print("سعر البيت:", hh[0])

#تحتاج حفظ للتدريب
```
